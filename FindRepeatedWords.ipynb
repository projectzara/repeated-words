```python
sample_file = open(working_dir + "/" + files_listing[0]) 
sample_file_lines = sample_file.readlines() 

sample_file_lines = sample_file_lines[0]

print(sample_file_lines)
```

```python
import hazm

normalizer = hazm.Normalizer()
```

```python
normalized_sample = normalizer.normalize(sample_file_lines)

print(normalized_sample)
```

```python
tokens = hazm.word_tokenize(normalized_sample)

print(tokens)
```

```python
def wordListToFreqDict(wordlist):
    wordfreq = [wordlist.count(p) for p in wordlist]
    return dict(list(zip(wordlist,wordfreq)))

repetition = wordListToFreqDict(tokens)
```
