```python
import json

json_file = open('words.json') 
json_file = json.load(json_file)

len(json_file)
```

```python
values = list(json_file.values()) 
keys = list(json_file.keys()) 

for value in values:
    position = values.index(value)
    print(f'{keys[position]} : {value}')
    
```

```python
for value in values:
    if value >= 250:
        position = values.index(value)
        print(f'{keys[position]} : {value}')
        
```

```python
must_delete = ".؟!)(۱۲۳۴۵۶۷۸۹-»«،:؛"
final_dict = {}

for value in values:
    if value >= 250:
        position = values.index(value)
        if keys[position] not in must_delete:
            final_dict[keys[position]] = value
print("done!")
```

```python
len(final_dict)
```

```python
output_file = open('most_frequent_words.txt', 'w') 

words = list(final_dict.keys())

for word in words:
    output_file.writeline(word)
    
output_file.close()
```
